## Linear simultaneous equations

- $$ x+3y=11 $$
- $$ 4x-5y=10 $$

```python
x,y=var('x y')
f1(x,y)= x+3*y
f2(x,y)= 4*x-5*y

print(f1(5,2),f2(5,2))
```

```python
4*f1(x,y)-f2(x,y)
```

```python
5*f1(x,y)+3*f2(x,y)
```

```python
x1, y1, x2, y2 = var('x1 y1 x2 y2')
eq1 = (x1-1)*(x1-2) + (y1-3)*(y1-4) == 0
eq2 = (x2-1)*(x2-2) + (y2-3)*(y2-4) == 0
eq3 = x1+x2 == 3
eq4 = y1+y2 == 7
sols = solve([eq1,eq2,eq3,eq4],x1,y1,x2,y2,solution_dict=True)
```

```python
print(sols)
```

## Simultaneous Equations with a Quadratic Equation

$$ x-y=7 $$
$$ y^2+xy+2x=5  $$

```python
x,y=var('x y')
f1(x,y)= x-y
f2(x,y)= y^2+x*y+2*x

print(f1(4,-3),f2(4,-3))

print(f1(5.5,-1.5),f2(5.5,-1.5))
```

```python
eq1 = f1(x,y) == 7
eq2 = f2(x,y) == 5

sols=solve([eq1,eq2],x,y,solution_dict=True)

print(sols)
```
