# Basics
Define a variable and a polynomial using the defined variable

```python
x=var('x')

f(x)=x^2-2*x-15

latex(f)
```

Factorise the polynomial and provide output in latex to obtain output in latex format note the double $$ around the latex expression

```python
f1=factor(f(x))
latex(f1)
```

The quadratic equation

$$x \ {\mapsto}\ x^{2} - 2 \, x - 15 $$

can be factorised as follows

$${\left(x + 3\right)} {\left(x - 5\right)}$$
