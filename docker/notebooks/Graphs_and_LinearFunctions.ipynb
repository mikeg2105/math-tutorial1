## Plotting a Graph

There is a very powerful way of visualising a function and this is to plot a diagram called a graph.

A graph two lines, which we call axes, these lines are at right angles to each other. Each line is a ruler or a number line. We call  the horizontal line the x-axis and the vertical line the y-axis.

Now, to plot a graph of our function we take an x value, input that value into the function and this will give a value for y.

This procees gives two numbers an x value and a y value, we call these values co-ordinates.

Our next task is to plot this point on our graph we measure along the x axis  our value for x, then moving out at right angles from the x-axis, we measure the y value along the y-axis.

We repeat this for a number of points, sage math provides the plot function which enables us to plot all the points which appear as a line.

```python
#The function definition
f(x)=x

#plot(f(x),(x,-15.0,15.0),axes_labels=['x axis','y axis'])
#show points on the plot as red diamond
show(point([(10,10), (8,8),(6,6),(4,4)],marker='d', markeredgecolor='red', size=20)+plot(f(x),(x,-15.0,15.0)),axes_labels=['x axis','y axis'])
```

We can see that the above plot is  a straight line and that for each value of x the corresponding value for y is always equal to x.

The sage plot function has a number of inputs, see we are using functions for our computer program. The function is like a box of tricks!

The first input for the sage plot function is the function definition.
The second input is the range of input values (x,-15,15), we can see the value x which we are going to vary the first number -15 is the starting value the second value is the final value for our plot.

### The Slope of a Straight Line

The steepness of our hill is also called the slope or the gradient. So if a hill is very steep we climb very high if we move along over a very short distance.

To calculate the slope we need at least two points or two pairs of coordinates (x1,y1) and (x2,y2).

Between these two points we move a distance of $$x_2-x_1$$ and our height increases by $$y_2-y_1$$

So, the slope is simply the height divided by the distance or
$${x_2-x_1}\over{y_2-y_1}$$

Try this calculation for our graph $$h(x)=f(x,5)= 5x+5$$

```python
print((h(10)-h(8))/(10-8))
print((h(10)-h(4))/(10-4))
print((h(8)-h(6))/(8-6))
print((h(6)-h(4))/(6-4))
```

Note this that we have the same value for the slope and amazingly this is the same as our value for m which was orignially 5!

What happens when we vary the slope, i.e. we change the value for m
